
••••••••••••••••ANÁLISIS EXPLORATORIO DEL MERCADO MINORISTA DE MIPYMES CUBANAS A PARTIR DE MÚLTIPLES FUENTES DE DATOS••••••••••••••••

En los últimos años, las mipymes han pasado a desempeñar un papel central en la comercialización minorista de productos básicos en Cuba. Sin embargo, el acceso real de la población a estos productos continúa estando limitado por factores como los precios, los métodos de pago disponibles y la inestabilidad del mercado informal de divisas.

Este proyecto analiza cómo se comportan los precios y la disponibilidad de productos esenciales en mipymes cubanas, y cómo estos precios se relacionan con la evolución del mercado informal de divisas, utilizando como caso de estudio el aceite comestible, uno de los productos de mayor demanda y sensibilidad económica.

El análisis se basa en datos reales recopilados directamente por mi y en fuentes digitales ampliamente utilizadas por la población cubana como grupos de ventas en redes sociales y revolico.

FUENTES DE DATOS UTILIZADAS: 

Para abordar la problemática se emplearon tres fuentes de datos independientes, lo que permite contrastar información desde distintos espacios del mercado:

1. Mipymes reales (visitadas físicamente o por sus grupos oficiales), dedicadas al comercio minorista, donde se recopilaron datos de 10 productos de alimentos y aseo en 30 empresas, incluyendo precios, disponibilidad y métodos de pago aceptados.

2. Análisis del precio del aceite en espacios digitales informales, específicamente grupos de Facebook, Telegram, Revolico y WhatsApp, que funcionan como mercados paralelos de referencia para la población.

3. Datos del medio ElToque, correspondientes al período comprendido entre el 1 de octubre y el 30 de noviembre, utilizados para analizar la evolución del precio informal del USD, el Euro y el MLC, así como su relación con el precio del aceite.

EL MERCADO INFORMAL DE DIVISAS Y SU EVOLUCIÓN

El mercado informal de divisas es un factor clave para entender la formación de precios en Cuba, especialmente en un contexto de alta dependencia de importaciones.

Este gráfico muestra la variación de los precios del USD y el Euro durante los meses de septiembre y octubre, según los datos publicados por El Toque.

La visualización evidencia la variación de las divisas y permite identificar tendencias que impactan directamente en los costos de reposición de mercancías en las mipymes.

```python
Graficar.graf_pl_ElToque()
```

En este grafico se puede apreciar fácilmente que tanto los precios del usd como el euro varian de la misma manera, es decir que ambos aumentan o disminuyen a la par y en la mayoria de los casos la misma cantidad, tambien evidencia claramente que el mercado informal de estas divisas es muy inestable presentando picos altos y caidas excesivas, el caso mas visible es específicamente entre el día 30 y el 39 es decir el 30 de octubre hasta el 8 de noviembre, durante este período ambas monedas sufrieron una caida excesiva en su precio disminuyendo aproximadamente un 15% el valor de ambas, algunos factores ocurridos en este período que pudieron influir en esta caida pudieran ser los siguientes:
-El exministro de Economía fue imputado por delitos graves, lo que generó incertidumbre y una caída en la demanda de divisas
-El gobierno anunció la futura reapertura de casas de cambio (CADECA) para vender divisas de forma oficial. Esta expectativa de oferta formal de divisas contribuyó a moderar la presión sobre el mercado informal
Estos cambios repentinos y frecuentes en la divisa sin lugar a duda afecta directamente tanto a la disponibilidad de los productos como a sus precios pues solo con estas monedas las mipymes pueden pagar las importaciones y no con la moneda nacional
