
••••••••••••••••ANÁLISIS EXPLORATORIO DEL MERCADO MINORISTA DE MIPYMES CUBANAS A PARTIR DE MÚLTIPLES FUENTES DE DATOS••••••••••••••••

En los últimos años, las mipymes han pasado a desempeñar un papel central en la comercialización minorista de productos básicos en Cuba. Sin embargo, el acceso real de la población a estos productos continúa estando limitado por factores como los precios, los métodos de pago disponibles y la inestabilidad del mercado informal de divisas.

Este proyecto analiza cómo se comportan los precios y la disponibilidad de productos esenciales en mipymes cubanas, y cómo estos precios se relacionan con la evolución del mercado informal de divisas, utilizando como caso de estudio el aceite comestible, uno de los productos de mayor demanda y sensibilidad económica.

El análisis se basa en datos reales recopilados directamente por mi y en fuentes digitales ampliamente utilizadas por la población cubana como grupos de ventas en redes sociales y revolico.

FUENTES DE DATOS UTILIZADAS: 

Para abordar la problemática se emplearon tres fuentes de datos independientes, lo que permite contrastar información desde distintos espacios del mercado:

1. Mipymes reales (visitadas físicamente o por sus grupos oficiales), dedicadas al comercio minorista, donde se recopilaron datos de 10 productos de alimentos y aseo en 30 empresas, incluyendo precios, disponibilidad y métodos de pago aceptados.

2. Análisis del precio del aceite en espacios digitales informales, específicamente grupos de Facebook, Telegram, Revolico y WhatsApp, que funcionan como mercados paralelos de referencia para la población.

3. Datos del medio ElToque, correspondientes al período comprendido entre el 1 de septiembre y el 31 de octubre, utilizados para analizar la evolución del precio informal del USD, el Euro y el MLC, así como su relación con el precio del aceite.
